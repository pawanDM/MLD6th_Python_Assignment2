```python
n=5
for i in range(n):
    for j in range(i):
        print ('* ', end= "")  # print generator always gives output in new line to overcome this we can use end parameter.
    print("")

for i in range(n,0,-1):
    for j in range(i):
        print('* ', end="")
    print("")
```

2. Write a Python program to reverse a word after accepting the input from the user.
Input word: ineuron
Output: norueni

```python
input_word = input("Enter a word :")

reverse_word = input_word [::-1]
print(reverse_word)
```
